```python
# Normalization at district level
import pandas as pd
import numpy as np

feature_cols = [
    'avg_temp_max', 'avg_temp_min', 'avg_humidity',
    'sum_precipitation', 'sum_sunshine_hours', 'mal_cases'
]

# Ensure required columns
assert 'district' in raw_data.columns and 'year' in raw_data.columns and 'month' in raw_data.columns

normalized_dfs = []
stats_per_district = {}

# Normalize per district
for district, df_district in raw_data.groupby('district'):
    df_district = df_district.sort_values(['year', 'month']).reset_index(drop=True)
    selected_data = df_district[feature_cols]
    data_values = selected_data.values.astype('float32')

    # Split for normalization
    num_samples = len(data_values)
    num_train = int(0.60 * num_samples)

    mean = data_values[:num_train].mean(axis=0)
    std = data_values[:num_train].std(axis=0)
    std[std < 1e-10] = 1.0  # avoid division by 0

    stats_per_district[district] = {'mean': mean, 'std': std}
    normalized_values = (data_values - mean) / std

    normalized_df = pd.DataFrame(normalized_values, columns=feature_cols, index=selected_data.index)
    normalized_df['district'] = district
    normalized_df['year'] = df_district['year'].values
    normalized_df['month'] = df_district['month'].values

    normalized_dfs.append(normalized_df)
```

### Step 4: Dataset Creation with District-Aware Splitting

```python
import numpy as np
import tensorflow as tf
from keras.utils import timeseries_dataset_from_array

def create_split_datasets(features, target, district_ids, sequence_length, delays, batch_size):
    max_delay = max(delays)
    total_offset = sequence_length + max_delay - 1

    train_datasets = []
    val_datasets = []
    test_datasets = []

    for district in np.unique(district_ids):
        mask = district_ids == district
        district_features = features[mask]
        district_target = target[mask]

        n = len(district_target)
        if n < total_offset:
            continue

        # Calculate the length of the shortest target array after applying the maximum delay
        min_length = n - max_delay
        # Create multi-step target by stacking arrays of the same length
        multi_targets = np.stack([district_target[d : min_length + d] for d in delays], axis=-1)
        total_sequences = len(multi_targets) - sequence_length + 1
        if total_sequences <= 0:
            continue

        train_seq = int(0.6 * total_sequences)
        val_seq = int(0.2 * total_sequences)
        test_seq = total_sequences - train_seq - val_seq

        def create_ds(start, end, shuffle):
            if end - start <= 0:
                return None
            return timeseries_dataset_from_array(
                data=district_features,
                targets=multi_targets[start:end],
                sequence_length=sequence_length,
                sampling_rate=1,
                batch_size=batch_size,
                shuffle=shuffle
            )

        ds_train = create_ds(0, train_seq, False)
        ds_val = create_ds(train_seq, train_seq + val_seq, False)
        ds_test = create_ds(train_seq + val_seq, train_seq + val_seq + test_seq, False)

        if ds_train: train_datasets.append(ds_train)
        if ds_val: val_datasets.append(ds_val)
        if ds_test: test_datasets.append(ds_test)

    def combine(datasets):
        if not datasets:  # Handle case where no datasets are created
            return None
        out = datasets[0]
        for ds in datasets[1:]:
            out = out.concatenate(ds)
        return out

    return combine(train_datasets), combine(val_datasets), combine(test_datasets)

# Call the function
train_ds, val_ds, test_ds = create_split_datasets(
    features, target, district_ids,
    sequence_length=6, delays=[1, 2, 3], batch_size=32
)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras import layers
from xgboost import XGBRegressor

# ====================
# CUSTOM LAYER
# ====================

class CastAndClipLayer(layers.Layer):
    """Custom layer to clip and cast district_ids to integers"""
    def __init__(self, num_districts, **kwargs):
        super().__init__(**kwargs)
        self.num_districts = num_districts

    def call(self, inputs):
        # Clip values to [0, num_districts-1] and cast to int32
        clipped = tf.clip_by_value(inputs, 0, self.num_districts - 1)
        return tf.cast(clipped, tf.int32)

    def get_config(self):
        config = super().get_config()
        config.update({"num_districts": self.num_districts})
        return config

# ====================
# 1. DATA PREPARATION FOR XGBOOST
# ====================

def prepare_xgboost_data(dataset, num_districts):
    """Convert TF dataset to numpy arrays for XGBoost, handling district_id"""
    X, y = [], []
    for batch in dataset:
        features, targets = batch
        # Split features into regular features and district_id
        regular_features = features[:, :, :-1]  # Shape: (batch_size, sequence_length, 5)
        district_ids = features[:, :, -1]      # Shape: (batch_size, sequence_length)
        # Convert to numpy, clip, and cast to integers
        district_ids_np = district_ids[:, -1].numpy()  # Shape: (batch_size,)
        district_ids_np = np.clip(district_ids_np, 0, num_districts - 1).astype(np.int32)
        # Debug: Print unique district_ids in batch
        print("XGBoost batch district_ids:", np.unique(district_ids_np))
        # One-hot encode district_ids
        one_hot_encoder = tf.keras.layers.CategoryEncoding(
            num_tokens=num_districts, output_mode="one_hot"
        )
        district_ids_encoded = one_hot_encoder(district_ids_np).numpy()  # Shape: (batch_size, num_districts)
        # Flatten regular features
        regular_features_flat = regular_features.numpy().reshape(features.shape[0], -1)  # Shape: (batch_size, sequence_length * 5)
        # Combine regular features and one-hot encoded district_ids
        combined_features = np.concatenate([regular_features_flat, district_ids_encoded], axis=-1)
        X.append(combined_features)
        y.append(targets.numpy())
    return np.concatenate(X), np.concatenate(y)

# ====================
# 2. MODEL ARCHITECTURES
# ====================

def dense_model(input_shape, num_districts):
    """Dense model with one-hot encoding for district_id"""
    inputs = keras.Input(shape=input_shape)  # (sequence_length=6, num_features=6)
    # Split input into features and district_id
    features = inputs[:, :, :-1]  # Shape: (batch_size, sequence_length, 5)
    district_ids = inputs[:, :, -1]  # Shape: (batch_size, sequence_length)
    # Select last time step and process with custom layer
    district_ids = district_ids[:, -1]  # Shape: (batch_size,)
    district_ids = CastAndClipLayer(num_districts)(district_ids)  # Shape: (batch_size,)
    # One-hot encode district_ids
    district_ids_encoded = layers.CategoryEncoding(
        num_tokens=num_districts, output_mode="one_hot"
    )(district_ids)  # Shape: (batch_size, num_districts)
    # Flatten features
    features_flat = layers.Reshape((input_shape[0] * (input_shape[1] - 1),))(features)  # Shape: (batch_size, sequence_length * 5)
    # Concatenate features and district_ids
    x = layers.Concatenate()([features_flat, district_ids_encoded])  # Shape: (batch_size, 30 + num_districts)
    x = layers.Dense(128, activation='relu')(x)
    x = layers.Dropout(0.2)(x)
    outputs = layers.Dense(3)(x)  # t+1, t+2, t+3
    return keras.Model(inputs, outputs)

def lstm_model(input_shape, num_districts):
    """LSTM model with one-hot encoding for district_id"""
    inputs = keras.Input(shape=input_shape)  # (sequence_length=6, num_features=6)
    features = inputs[:, :, :-1]  # Shape: (batch_size, sequence_length, 5)
    district_ids = inputs[:, :, -1]  # Shape: (batch_size, sequence_length)
    district_ids = district_ids[:, -1]  # Shape: (batch_size,)
    district_ids = CastAndClipLayer(num_districts)(district_ids)  # Shape: (batch_size,)
    district_ids_encoded = layers.CategoryEncoding(
        num_tokens=num_districts, output_mode="one_hot"
    )(district_ids)  # Shape: (batch_size, num_districts)
    x = layers.LSTM(128)(features)
    x = layers.Dropout(0.5)(x)
    x = layers.Concatenate()([x, district_ids_encoded])
    outputs = layers.Dense(3)(x)
    return keras.Model(inputs, outputs)

def gru_model(input_shape, num_districts):
    """GRU model with one-hot encoding for district_id"""
    inputs = keras.Input(shape=input_shape)  # (sequence_length=6, num_features=6)
    features = inputs[:, :, :-1]  # Shape: (batch_size, sequence_length, 5)
    district_ids = inputs[:, :, -1]  # Shape: (batch_size, sequence_length)
    district_ids = district_ids[:, -1]  # Shape: (batch_size,)
    district_ids = CastAndClipLayer(num_districts)(district_ids)  # Shape: (batch_size,)
    district_ids_encoded = layers.CategoryEncoding(
        num_tokens=num_districts, output_mode="one_hot"
    )(district_ids)  # Shape: (batch_size, num_districts)
    x = layers.GRU(256)(features)
    x = layers.Concatenate()([x, district_ids_encoded])
    outputs = layers.Dense(3)(x)
    return keras.Model(inputs, outputs)

def transformer_model(input_shape, num_districts, head_size=64, num_heads=6, ff_dim=128, num_transformer_blocks=4):
    """Transformer model with one-hot encoding for district_id"""
    inputs = keras.Input(shape=input_shape)  # (sequence_length=6, num_features=6)
    features = inputs[:, :, :-1]  # Shape: (batch_size, sequence_length, 5)
    district_ids = inputs[:, :, -1]  # Shape: (batch_size, sequence_length)
    district_ids = district_ids[:, -1]  # Shape: (batch_size,)
    district_ids = CastAndClipLayer(num_districts)(district_ids)  # Shape: (batch_size,)
    district_ids_encoded = layers.CategoryEncoding(
        num_tokens=num_districts, output_mode="one_hot"
    )(district_ids)  # Shape: (batch_size, num_districts)
    x = layers.LayerNormalization(epsilon=1e-6)(features)
    for _ in range(num_transformer_blocks):
        attention_output = layers.MultiHeadAttention(
            key_dim=head_size,
            num_heads=num_heads,
            dropout=0.2
        )(x, x)
        x = layers.Add()([x, attention_output])
        x = layers.LayerNormalization(epsilon=1e-6)(x)
        ffn = layers.Dense(ff_dim, activation="relu")(x)
        ffn = layers.Dense(features.shape[-1])(ffn)  # Match feature dimension (5)
        x = layers.Add()([x, ffn])
        x = layers.LayerNormalization(epsilon=1e-6)(x)
    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Concatenate()([x, district_ids_encoded])
    outputs = layers.Dense(3)(x)
    return keras.Model(inputs, outputs)

# ====================
# 3. TRAINING PIPELINE
# ====================

def compile_and_train(model, train_dataset, val_dataset, name, epochs=500, patience=20):
    """Universal training function"""
    model.compile(optimizer='rmsprop', loss='mae', metrics=['mae'])
    print(f"\nTraining {name}...")
    callbacks = [
        keras.callbacks.EarlyStopping(patience=patience, restore_best_weights=True),
        keras.callbacks.ReduceLROnPlateau(factor=0.5, patience=5)
    ]
    history = model.fit(
        train_dataset,
        validation_data=val_dataset,
        epochs=epochs,
        callbacks=callbacks,
        verbose=1
    )
    return model, history

def train_xgboost(X_train, y_train, X_val, y_val):
    """XGBoost training with early stopping"""
    model = XGBRegressor(
        n_estimators=1000,
        learning_rate=0.001,
        objective='reg:squarederror',
        early_stopping_rounds=100,
        eval_metric='mae'
    )
    print("\nTraining XGBoost...")
    model.fit(
        X_train, y_train,
        eval_set=[(X_val, y_val)],
        verbose=1
    )
    return model

def train_models(train_dataset, val_dataset, num_districts):
    """Train all models with district-aware data"""
    input_shape = next(iter(train_dataset))[0].shape[1:]  # (sequence_length=6, num_features=6)
    print(f"Input shape: {input_shape}")
    models = {
        'dense': dense_model(input_shape, num_districts),
        'LSTM': lstm_model(input_shape, num_districts),
        'GRU': gru_model(input_shape, num_districts),
        'transformer': transformer_model(input_shape, num_districts)
    }
    histories = {}
    for name, model in models.items():
        models[name], histories[name] = compile_and_train(
            model, train_dataset, val_dataset, name
        )
    # Train XGBoost
    X_train_flat, y_train_flat = prepare_xgboost_data(train_dataset, num_districts)
    X_val_flat, y_val_flat = prepare_xgboost_data(val_dataset, num_districts)
    models['XGBoost'] = train_xgboost(X_train_flat, y_train_flat, X_val_flat, y_val_flat)
    return {
        'models': models,
        'histories': histories
    }

# ====================
# 4. EXECUTE TRAINING
# ====================

# Verify district_id values
unique_districts = set()
for X, _ in train_ds:
    unique_districts.update(X[:, :, -1].numpy().flatten())
print("Unique district IDs:", sorted(unique_districts))
num_districts = len(unique_districts)  # Should be 118

# Example usage
training_output = train_models(train_ds, val_ds, num_districts)
```

```python
import numpy as np
import tensorflow as tf

def compute_mae_per_timestep(models_dict, test_dataset, num_districts, dataset_name="Test"):
    """
    Compute average MAE across all districts for each time step (t+1, t+2, t+3).
    
    Args:
        models_dict: Dictionary of trained models (e.g., training_output['models'])
        test_dataset: TensorFlow dataset (e.g., test_ds) with (X, y) batches
        num_districts: Number of districts (e.g., 118)
        dataset_name: Name of dataset for printing (e.g., "Test" or "Validation")
    
    Returns:
        Dictionary with MAE per time step for each model
    """
    # Initialize MAE accumulators for each model and time step
    mae_per_timestep = {name: np.zeros(3) for name in models_dict}
    sample_counts = {name: 0 for name in models_dict}
    
    # Process dataset batches
    for X, y in test_dataset:
        district_ids = X[:, -1, -1].numpy().astype(int)  # Last time step's district_id
        y_true = y.numpy()  # Shape: (batch_size, 3)
        
        for model_name, model in models_dict.items():
            if model_name != 'XGBoost':
                # Keras model predictions
                y_pred = model.predict(X, verbose=0)  # Shape: (batch_size, 3)
            else:
                # XGBoost predictions
                X_flat, _ = prepare_xgboost_data(X, num_districts)  # Convert batch
                y_pred = model.predict(X_flat)  # Shape: (batch_size, 3)
            
            # Compute absolute errors
            errors = np.abs(y_pred - y_true)  # Shape: (batch_size, 3)
            # Accumulate errors for each time step
            mae_per_timestep[model_name] += np.sum(errors, axis=0)  # Sum over batch
            sample_counts[model_name] += errors.shape[0]  # Count samples
    
    # Compute average MAE per time step
    for model_name in mae_per_timestep:
        mae_per_timestep[model_name] /= sample_counts[model_name]
        print(f"\n{model_name} {dataset_name} MAE per Time Step:")
        for t in range(3):
            print(f"  t+{t+1}: {mae_per_timestep[model_name][t]:.4f}")
    
    return mae_per_timestep

# Helper function for XGBoost (from your code, adapted for single batch)
def prepare_xgboost_data(batch_features, num_districts):
    """Convert a single batch to NumPy arrays for XGBoost"""
    # Ensure batch_features is a tensor
    if isinstance(batch_features, tuple):
        batch_features = batch_features[0]  # Extract features if dataset yields (X, y)
    
    # Split features into regular features and district_id
    regular_features = batch_features[:, :, :-1]  # Shape: (batch_size, sequence_length, 5)
    district_ids = batch_features[:, :, -1]       # Shape: (batch_size, sequence_length)
    # Convert to NumPy, clip, and cast to integers
    district_ids_np = district_ids[:, -1].numpy()  # Shape: (batch_size,)
    district_ids_np = np.clip(district_ids_np, 0, num_districts - 1).astype(np.int32)
    # One-hot encode district_ids
    one_hot_encoder = tf.keras.layers.CategoryEncoding(
        num_tokens=num_districts, output_mode="one_hot"
    )
    district_ids_encoded = one_hot_encoder(district_ids_np).numpy()  # Shape: (batch_size, num_districts)
    # Flatten regular features
    regular_features_flat = regular_features.numpy().reshape(batch_features.shape[0], -1)  # Shape: (batch_size, sequence_length * 5)
    # Combine features
    combined_features = np.concatenate([regular_features_flat, district_ids_encoded], axis=-1)
    return combined_features, None  # Return None for y since not needed

# Example usage
num_districts = 118  # From your unique_districts output
mae_results = compute_mae_per_timestep(training_output['models'], test_ds, num_districts, dataset_name="Test")
```

```python
import pandas as pd
import numpy as np

def export_test_maes(mae_results, stats_per_district, filename='test_maes.csv'):
    """
    Export denormalized test MAEs for each model to a CSV file.
    
    Args:
        mae_results: Dictionary with model names as keys and MAE arrays (shape: (3,))
        stats_per_district: Dictionary with district-specific mean and std for features
        filename: Output CSV filename
    """
    output_data = []
    
    for model_name, mae_values in mae_results.items():
        # Calculate average MAE across all districts
        district_maes = []
        for district in stats_per_district:
            mal_cases_idx = -1  # mal_cases is the last feature
            mean = stats_per_district[district]['mean'][mal_cases_idx]
            std = stats_per_district[district]['std'][mal_cases_idx]
            if std < 1e-10:
                std = 1.0  # Avoid division by zero
            # Denormalize MAE for this district
            denorm_mae = mae_values * std
            district_maes.append(denorm_mae)
        
        # Average MAE across districts
        avg_mae_per_timestep = np.mean(district_maes, axis=0)
        
        # Calculate sum and average MAE
        mae_sum = np.sum(avg_mae_per_timestep)
        mae_avg = np.mean(avg_mae_per_timestep)
        
        output_data.append({
            'model_name': model_name,
            'mae_sum': mae_sum,
            'mae': mae_avg,
            't+1_mae': avg_mae_per_timestep[0],
            't+2_mae': avg_mae_per_timestep[1],
            't+3_mae': avg_mae_per_timestep[2]
        })
    
    # Create DataFrame and save to CSV
    output_df = pd.DataFrame(output_data)
    output_df.to_csv(filename, index=False)
    print(f"Test MAEs saved to {filename}")

# Example usage
export_test_maes(mae_results, stats_per_district)
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from datetime import datetime

# Step 1: Remove Kamuli and create district-to-ID mapping
monthly_data = monthly_data[monthly_data['district'] != 'Kamuli'].copy()
unique_districts = sorted(monthly_data['district'].unique())  # Should be 118 districts
district_to_id = {district: idx for idx, district in enumerate(unique_districts)}
num_districts = len(unique_districts)  # Should be 118
print(f"Number of districts after removing Kamuli: {num_districts}")

# Step 2: Prepare input data with district-level normalization
def prepare_input_data(df, stats_per_district, start_date='2024-11', end_date='2025-04'):
    """
    Prepare input data for prediction, normalizing features per district and skipping incomplete data.
    
    Args:
        df: pandas DataFrame with columns [year, month, district, mal_cases, avg_temp_max, 
            avg_temp_min, avg_humidity, sum_precipitation, sum_sunshine_hours]
        stats_per_district: dict with district-specific mean and std for features
        start_date: str, start of the window (e.g., '2024-11')
        end_date: str, end of the window (e.g., '2025-04')
    
    Returns:
        tuple: (input_data, valid_districts, valid_indices)
            - input_data: numpy array of shape (num_valid_districts, 6, 6)
            - valid_districts: list of district names with complete data
            - valid_indices: list of district IDs for valid districts
    """
    # Convert year and month to a datetime for filtering
    df['date'] = pd.to_datetime(df[['year', 'month']].assign(day=1))
    start = pd.to_datetime(start_date + '-01')
    end = pd.to_datetime(end_date + '-01')
    
    # Filter data for the specified window
    mask = (df['date'] >= start) & (df['date'] <= end)
    input_df = df[mask].copy()
    
    # Check for 6 months
    expected_months = pd.date_range(start, end, freq='MS')
    if len(input_df['date'].unique()) < 6:
        print(f"Warning: Found {len(input_df['date'].unique())} months in dataset, expected 6")
    
    # Initialize lists for valid districts
    input_data = []
    valid_districts = []
    valid_indices = []
    
    # Features to normalize
    feature_cols = ['avg_temp_max', 'avg_temp_min', 'avg_humidity', 'sum_precipitation', 'sum_sunshine_hours']
    
    for district_idx, district in enumerate(unique_districts):
        # Filter data for the district
        district_data = input_df[input_df['district'] == district].sort_values('date')
        if len(district_data) != 6:
            print(f"Warning: District {district} has {len(district_data)} months, expected 6. Skipping.")
            continue
        
        # Extract features
        district_features = district_data[feature_cols].to_numpy()  # Shape: (6, 5)
        
        # Normalize features using district-specific stats
        if district not in stats_per_district:
            print(f"Warning: No normalization stats for {district}. Skipping.")
            continue
        mean = stats_per_district[district]['mean'][:-1]  # Exclude mal_cases
        std = stats_per_district[district]['std'][:-1]    # Exclude mal_cases
        std[std < 1e-10] = 1.0  # Avoid division by zero
        normalized_features = (district_features - mean) / std  # Shape: (6, 5)
        
        # Add district_id as the last feature (not normalized)
        district_id = np.full((6, 1), district_to_id[district])
        # Combine features and district_id
        district_input = np.concatenate([normalized_features, district_id], axis=-1)  # Shape: (6, 6)
        input_data.append(district_input)
        valid_districts.append(district)
        valid_indices.append(district_idx)
    
    if not input_data:
        raise ValueError("No districts have complete data for the specified window.")
    
    return np.array(input_data), valid_districts, valid_indices

# Step 3: Prepare data for XGBoost
def prepare_xgboost_data(batch_features, num_districts):
    """
    Convert a batch of features to NumPy arrays for XGBoost.
    
    Args:
        batch_features: numpy array of shape (batch_size, sequence_length, 6)
        num_districts: int, number of districts (118)
    
    Returns:
        tuple: (combined_features, None)
            - combined_features: numpy array of shape (batch_size, sequence_length * 5 + num_districts)
            - None: placeholder for y (not needed for prediction)
    """
    # Ensure batch_features is a NumPy array
    if isinstance(batch_features, tf.Tensor):
        batch_features = batch_features.numpy()
    
    # Split features into regular features and district_id
    regular_features = batch_features[:, :, :-1]  # Shape: (batch_size, sequence_length, 5)
    district_ids = batch_features[:, :, -1]       # Shape: (batch_size, sequence_length)
    
    # Extract district_id from the last time step
    district_ids_np = district_ids[:, -1]  # Shape: (batch_size,)
    district_ids_np = np.clip(district_ids_np, 0, num_districts - 1).astype(np.int32)
    
    # One-hot encode district_ids
    one_hot_encoder = tf.keras.layers.CategoryEncoding(
        num_tokens=num_districts, output_mode="one_hot"
    )
    district_ids_encoded = one_hot_encoder(district_ids_np).numpy()  # Shape: (batch_size, num_districts)
    
    # Flatten regular features
    regular_features_flat = regular_features.reshape(batch_features.shape[0], -1)  # Shape: (batch_size, sequence_length * 5)
    
    # Combine features
    combined_features = np.concatenate([regular_features_flat, district_ids_encoded], axis=-1)
    return combined_features, None

# Step 4: Make predictions and denormalize
def make_predictions(models_dict, input_data, valid_districts, stats_per_district):
    """
    Generate and denormalize predictions for May, June, July 2025 for each district.
    
    Args:
        models_dict: Dictionary of trained models
        input_data: numpy array of shape (num_valid_districts, 6, 6)
        valid_districts: list of valid district names
        stats_per_district: dict with district-specific mean and std for features
    
    Returns:
        Dictionary with denormalized predictions for each model, shape (num_valid_districts, 3)
    """
    predictions = {name: [] for name in models_dict}
    
    # Batch predictions
    for model_name, model in models_dict.items():
        if model_name != 'XGBoost':
            # Keras model predictions
            y_pred = model.predict(input_data, verbose=0)  # Shape: (num_valid_districts, 3)
        else:
            # XGBoost predictions
            X_flat, _ = prepare_xgboost_data(input_data, num_districts)  # Shape: (num_valid_districts, 30 + num_districts)
            print(f"XGBoost input shape: {X_flat.shape}")  # Debug: Verify shape
            y_pred = model.predict(X_flat)  # Shape: (num_valid_districts, 3)
        
        # Denormalize predictions (mal_cases only)
        denorm_preds = np.zeros_like(y_pred)  # Shape: (num_valid_districts, 3)
        for i, district in enumerate(valid_districts):
            mal_cases_idx = -1  # mal_cases is the last feature in stats_per_district
            mean = stats_per_district[district]['mean'][mal_cases_idx]
            std = stats_per_district[district]['std'][mal_cases_idx]
            if std < 1e-10:
                std = 1.0  # Avoid division by zero
            denorm_preds[i] = y_pred[i] * std + mean
        
        predictions[model_name] = denorm_preds
    
    return predictions

# Step 5: Export predictions to CSV
def export_predictions(predictions, districts, filename='predictions_may_june_july_2025.csv'):
    """
    Export predictions to a CSV file.
    
    Args:
        predictions: Dictionary with model names as keys and arrays of shape (num_valid_districts, 3)
        districts: List of valid district names
        filename: Output CSV filename
    """
    output_data = []
    months = ['May_2025', 'June_2025', 'July_2025']
    
    for district_idx, district in enumerate(districts):
        for model_name in predictions:
            preds = predictions[model_name][district_idx]
            for month_idx, month in enumerate(months):
                output_data.append({
                    'district': district,
                    'model': model_name,
                    'month': month,
                    'predicted_mal_cases': preds[month_idx]
                })
    
    # Create DataFrame and save to CSV
    output_df = pd.DataFrame(output_data)
    output_df.to_csv(filename, index=False)
    print(f"Predictions saved to {filename}")

# Example usage
try:
    input_data, valid_districts, valid_indices = prepare_input_data(monthly_data, stats_per_district, start_date='2024-11', end_date='2025-04')
    print(f"Valid districts for prediction: {len(valid_districts)} out of {num_districts}")
    predictions = make_predictions(training_output['models'], input_data, valid_districts, stats_per_district)
    export_predictions(predictions, valid_districts)
    
    # Optional: Print sample predictions for the first valid district
    if valid_districts:
        for model_name in predictions:
            print(f"\n{model_name} Predictions for {valid_districts[0]}:")
            print(f"  May 2025: {predictions[model_name][0, 0]:.2f}")
            print(f"  June 2025: {predictions[model_name][0, 1]:.2f}")
            print(f"  July 2025: {predictions[model_name][0, 2]:.2f}")
except ValueError as e:
    print(f"Error: {e}")
```
